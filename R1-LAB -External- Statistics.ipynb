```python
# Load Packages
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read data set and display top 10 rows
df = pd.read_csv('Laliga.csv', header=1)
df.head(10)
```

## 1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (5 points)

```python
# replace all '-' with '0'
df.replace(to_replace='-',value=0,inplace=True)
df.head(10)
```

```python
df.info()
```

## 2. Print all the teams which have started playing between 1930-1980. (5 points)

```python
# Columns 'Debut' has Object type data hence cannot be used for numerical calculation i.e 1930-31
# created a new colum 'Debut No' in which the first part of year will be stored and will be converted to Int64 i.e. 1930
df['Debut_No'] = df['Debut'] # create a new column 'debut_no'
row = 0
for rows in df['Debut_No']:
    df.loc[row, 'Debut_No'] = rows[:4]
    row +=1
df['Debut_No'] =  df['Debut_No'].astype('int64')
```

```python
# Now it is easy to find the team that started playing between 2 dates 1930 and 1980
df[(df['Debut_No']>=1930) & (df['Debut_No']<1980)]
```

## 3. Print the list of teams which came Top 5 in terms of points (5 points)

```python
# Change data type of 'Points' to Int64
df['Points'] = df['Points'].astype('int64')
```

```python
# Print top 5 rows after sorting the data by 'Points' in Descening order
df.sort_values(by = 'Points', ascending = False)[:5]
```

## 4. Write a function with name Goal_diff_count which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

###### Goal_diff_count = GoalsFor - GoalsAgainst

```python
def Goal_diff_count():
    Goal_diff_count = df['GoalsFor'].astype('int64')- df['GoalsAgainst'].astype('int64')
    return (Goal_diff_count) # We can also return absloute value to show actual difference and not in -ve
```

```python
goal_diff = list(Goal_diff_count())
```

```python
max_goal_diff_index = goal_diff.index(max(goal_diff))
min_goal_diff_index = goal_diff.index(min(goal_diff))
print ("'{0}' has scored {1} Goals 'For' and {2} Goals 'Against' which is Maximum Difference of {3} \n"
       .format(df.loc[max_goal_diff_index,"Team"],df.loc[max_goal_diff_index,"GoalsFor"],df.loc[max_goal_diff_index,"GoalsAgainst"],max(goal_diff)))
print ("'{0}' has scored {1} Goals 'For' and {2} Goals 'Against' which is Minimum Difference of {3} \n"
       .format(df.loc[min_goal_diff_index,"Team"],df.loc[min_goal_diff_index,"GoalsFor"],df.loc[min_goal_diff_index,"GoalsAgainst"],min(goal_diff)))
```

## 5. Create a new column with name Winning Percent and append it to the data set (5 points)

#### Percentage of Winning = (GamesWon / GamesPlayed)*100

```python
# Create new column ' Winning_percent'
df['Winning_percent'] = df['GamesWon'].astype('int64')/df['GamesPlayed'].astype('int64')/100
```

#### If there are any numerical error, replace it with 0%

```python
# Replace NAN values with 0
df['Winning_percent'].replace(to_replace=np.NAN,value=0,inplace=True)
```

```python
# Plotting histogram successfully shows there are no error values
plt.hist(df['Winning_percent'])
plt.show()
```

#### Print the top 5 teams which has the highest Winning percentage

```python
# Display top 5 after sorting the data frame in decsnbing order with Winning Percentange
df.sort_values(by='Winning_percent',ascending = False).head(5)
```
